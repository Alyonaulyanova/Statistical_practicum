```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import pearsonr, f_oneway

plt.style.use('default')
sns.set(style="whitegrid")

np.random.seed(42)


from google.colab import drive
drive.mount('/content/drive')

path = '/content/drive/MyDrive/Colab Notebooks/IMDB-Movie-Data.csv.xls'

df = pd.read_csv(path)

data = df.copy()
data.head()

print("\n Размер датасета")
print(data.shape)

print("\n Первые строки")
display(data.head())

print("\n Информация")
data.info()

print("\n Описательная статистика")
display(data.describe())


print("\n Типы данных")
print(data.dtypes)

missing = data.isnull().sum().sort_values(ascending=False)
missing_percent = (missing / len(data)) * 100

missing_df = pd.DataFrame({
    'Missing': missing,
    'Percent': missing_percent
})

print("\n Пропуски")
display(missing_df)

plt.figure(figsize=(10,5))
sns.barplot(x=missing_df.index, y=missing_df['Percent'])
plt.xticks(rotation=45)
plt.title('Процент пропусков по признакам')
plt.xlabel('Признаки')
plt.ylabel('Процент пропусков')
plt.show()


plt.figure(figsize=(8,5))
sns.histplot(data['Rating'], bins=20, kde=True)
plt.title('Распределение рейтинга')
plt.xlabel('Rating')
plt.ylabel('Количество')
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x=data['Rating'])
plt.title('Boxplot рейтинга')
plt.show()


numeric_cols = data.select_dtypes(include=['int64', 'float64']).columns

print("\n Числовые признаки")
print(numeric_cols)

display(data[numeric_cols].describe())

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(data[col], bins=30, kde=True)
    plt.title(f'Распределение {col}')
    plt.xlabel(col)
    plt.ylabel('Количество')
    plt.show()

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot {col}')
    plt.show()


if 'Votes' in data.columns:
    data['Log_Votes'] = np.log1p(data['Votes'])

    plt.figure(figsize=(6,4))
    sns.histplot(data['Log_Votes'], kde=True)
    plt.title('Log(Votes)')
    plt.xlabel('Log_Votes')
    plt.ylabel('Количество')
    plt.show()

categorical_cols = data.select_dtypes(include=['object']).columns

print("\n Категориальные признаки")
print(categorical_cols)

for col in categorical_cols:
    print(f"\n {col}")
    print(data[col].value_counts().head(10))


if 'Director' in data.columns:
    top_directors = data['Director'].value_counts().head(10)

    plt.figure(figsize=(8,5))
    sns.barplot(x=top_directors.values, y=top_directors.index)
    plt.title('Топ режиссеров')
    plt.xlabel('Количество фильмов')
    plt.ylabel('Режиссер')
    plt.show()

if 'Actors' in data.columns:
    data['Actors_list'] = data['Actors'].apply(lambda x: x.split(', '))
    data['Actors_count'] = data['Actors_list'].apply(len)

if 'Genre' in data.columns:
    data['Genre_list'] = data['Genre'].apply(lambda x: x.split(', '))
    data['Genre_count'] = data['Genre_list'].apply(len)

if 'Director' in data.columns:
    director_counts = data['Director'].value_counts()
    data['Director_freq'] = data['Director'].map(director_counts)


plt.figure(figsize=(10,6))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица')
plt.show()


if 'Votes' in data.columns:
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=data['Votes'], y=data['Rating'])
    plt.title('Votes vs Rating')
    plt.xlabel('Votes')
    plt.ylabel('Rating')
    plt.show()

if 'Log_Votes' in data.columns:
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=data['Log_Votes'], y=data['Rating'])
    plt.title('Log(Votes) vs Rating')
    plt.xlabel('Log_Votes')
    plt.ylabel('Rating')
    plt.show()

if 'Genre' in data.columns:
    top_genres = data['Genre'].value_counts().head(10).index
    subset = data[data['Genre'].isin(top_genres)]

    plt.figure(figsize=(10,5))
    sns.boxplot(x='Genre', y='Rating', data=subset)
    plt.xticks(rotation=45)
    plt.title('Жанры vs Rating')
    plt.show()



def detect_outliers(col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = data[(data[col] < Q1 - 1.5*IQR) | (data[col] > Q3 + 1.5*IQR)]
    return len(outliers)

print("\n Выбросы")
for col in numeric_cols:
    print(col, detect_outliers(col))

if 'Votes' in data.columns:
    corr, p = pearsonr(data['Votes'], data['Rating'])
    print(f'\nPearson correlation (Votes vs Rating): {corr:.3f}, p-value: {p:.5f}')

if 'Genre' in data.columns:
    groups = [group['Rating'].values for name, group in data.groupby('Genre') if len(group) > 5]

    if len(groups) >= 2:
        stat, p = f_oneway(*groups[:5])
        print(f'ANOVA p-value (Genre vs Rating): {p:.5f}')


data.to_csv('processed_imdb.csv', index=False)
```
