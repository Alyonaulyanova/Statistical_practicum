```python
def compute_sum(start, end, step):
    total = 0.0
    i = start
    while (step > 0 and i <= end) or (step < 0 and i >= end):
        total += 1.0 / (i * i)  # добавляем 1/i^2
        i += step
    return total

A = compute_sum(1, 10000, 1)

B = compute_sum(10000, 1, -1)

print(f"A (от 1 до 10000) = {A:.15f}")
print(f"B (от 10000 до 1) = {B:.15f}")
print(f"A - B = {A - B:.15e}")
```

Объяснение:
из-за конечной точности чисел с плавающей точкой (double precision) порядок суммирования влияет на результат. При суммировании от малых к большим (A) маленькие числа теряются,когда они добавляются к уже большой сумме. При суммировании от больших к малым (B) сначала складываются маленькие числа, что даёт более точный результат.

Лучший способ:
суммировать от меньших чисел к большим (по убыванию индекса),
чтобы сначала накапливались маленькие величины и ошибка округления была меньше.
